```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataframe=pd.read_csv('data.csv')
```

```python
dataframe
```

```python
redefined=dataframe.drop(['index','price','sq_price'],axis=1)[0:10]
```

```python
redefined
```

```python
#add labels 1=good buy; 0=bad buy
redefined.loc[:,('y1')]=[1,1,0,1,0,0,1,1,1,1]
redefined.loc[:,('y2')]=redefined['y1']==0
redefined.loc[:,('y2')]=redefined['y2'].astype(int)
```

```python
redefined
```

```python
# giddy up data for tensorflow
#convert features to tensors
inputData=redefined.loc[:,['area','bathrooms']].as_matrix()
#convert labels to tensors
labelData=redefined.loc[:,['y1','y2']].as_matrix()
```

```python
#create hyperparamters
learningRate=0.0001 #- how fast do we get to convergence
batchSize=10 #set of training data used for training
epochs=2000 #how many times to run
display_step=50 #how often to display
nsamples=labelData.size
```

```python
#create computation graph/neural network
# '2' features, None- means any no of examples
# placeholders are gateways
x=tf.placeholder(tf.float32,[None,2])

#create weights
W=tf.Variable(tf.zeros([2,2]))
#add biases, help model fit better
b=tf.Variable(tf.zeros([2]))

#multiply weights with biases
y_values=tf.add(tf.matmul(x,W),b)

#apply softmax to value just created, this is activation function
y=tf.nn.softmax(y_values)

y_=tf.placeholder(tf.float32,[None,2])
```

```python
#perform training
#reduce sum, computes sum of elements across dimensions of tensor
cost=tf.reduce_mean(tf.pow(y_ - y,2))/ (2*nsamples)
#gradient descent
optimizer=tf.train.GradientDescentOptimizer(learningRate).minimize(cost)

```

```python
#initialize variables and tensorflow session
init=tf.initialize_all_variables()
sess=tf.Session()
sess.run(init)
```

```python
for i in range(epochs):  
    sess.run(optimizer, feed_dict={x: inputData, y_: labelData}) # Take a gradient descent step using our inputs and labels

    # That's all! The rest of the cell just outputs debug messages. 
    # Display logs per epoch step
    if (i) % display_step == 0:
        cc = sess.run(cost, feed_dict={x: inputData, y_:labelData})
        print ('Training step: '+'%04d'%i+ 'cost= {:.9f}'.format(cc)) 

print ("Optimization Finished!")
training_cost = sess.run(cost, feed_dict={x: inputData, y_: labelData})
print ('Training cost='+ str(training_cost))
print('W='+ str(sess.run(W)))
print('b='+ str(sess.run(b)))
```
